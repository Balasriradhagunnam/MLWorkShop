```python
plt.hist(data['Subscribers (millions)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Subscribers (in millions)', fontsize=15)
plt.xlabel('Subscribers (millions)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

This histogram shows the distribution of the number of subscribers across all the YouTube channels in the dataset.

The majority of channels fall in the lower range of subscribers, indicating that the top channels with massive subscribers are outliers compared to the rest.

```python
# Total subscribers per category
category_subscribers = data.groupby('Category')['Subscribers (millions)'].sum().sort_values(ascending=False)
# Plot
category_subscribers.plot(kind='bar', color='orange')
plt.title('Total Subscribers per Category', fontsize=15)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Subscribers (millions)', fontsize=12)
plt.show()
```


This bar plot visualizes the total number of subscribers per category.

Entertainment holds the highest total number of subscribers, followed by Music and Education.

 These categories dominate the platform, likely because of the widespread popularity of entertainment-related content.

Smaller categories, such as Sports, Film, and How-to, have much lower subscriber counts, indicating that while they may have dedicated audiences, they are not as globally popular as entertainment or music channels.

```python
# Total subscribers per country
country_subscribers = data.groupby('Country')['Subscribers (millions)'].sum().sort_values(ascending=False)
country_subscribers.plot(kind='bar', color='green')
plt.title('Total Subscribers per Country', fontsize=15)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Subscribers (millions)', fontsize=12)
plt.show()
```

The United States stands out with the highest total subscriber count,countries with smaller populations or less representation, like Cyprus and Belarus, have significantly fewer total subscribers.

```python
# Boxplot to compare subscriber distribution based on 'Brand channel'
sns.boxplot(x='Brand channel', y='Subscribers (millions)', data=data, palette='Set2')
plt.title('Subscriber Distribution by Brand Channel', fontsize=15)
plt.xlabel('Brand Channel', fontsize=12)
plt.ylabel('Subscribers (millions)', fontsize=12)
plt.show()
```

This boxplot compares the distribution of subscribers for channels with and without a Brand Channel.

Brand channels (those with a "Yes" in the "Brand channel" column) show a wider range of subscribers with some very high subscriber counts (like T-Series and MrBeast).

Non-brand channels (those with a "No" in the "Brand channel" column) show a much narrower range of subscribers, with most channels moving around the lower subscriber ranges.

The upper whisker for brand channels extends much farther, indicating that these channels include a few extremely popular creators, while non-brand channels tend to have smaller subscriber bases.

```python
# Correlation matrix
corr_matrix = data[['Subscribers (millions)']].corr()
sns.heatmap(corr_matrix, annot=True, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

Since there’s only one numerical variable (Subscribers), the correlation matrix will only show the self-correlation because the correlation of subscribers with itself will be 1.
